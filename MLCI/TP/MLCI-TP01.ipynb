```python
import numpy
```

```python
D1=numpy.random.rand(100,2)
D1.shape
```

```python
D2=4*numpy.random.rand(100,2)
D2.shape
```

```python
D3=2*numpy.random.rand(100,2)
D3.shape
```

```python
X1 = D1[:, 0].reshape(-1, 1)
X2 = (D2[:, 0] + 6).reshape(-1, 1)
X3 = (D3[:, 0] + 8).reshape(-1, 1)

X = numpy.concatenate((X1, X2, X3), axis=0)
print("X shape:", X.shape)
```

```python
Y1 = D1[:, 1].reshape(-1, 1)
Y2 = (D2[:, 1] + 6).reshape(-1, 1)
Y3 = (D3[:, 1] + 3).reshape(-1, 1)

Y = numpy.concatenate((Y1, Y2, Y3), axis=0)
print("Y shape:", Y.shape)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y, c='g', marker='+')
plt.show()
```

```python
from sklearn.cluster import KMeans
import numpy as np


# Stack X and Y to form the complete dataset (300x2 matrix)
dataset = np.hstack((X, Y))

# Apply K-means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(dataset)

# Get the cluster centroids
centroids = kmeans.cluster_centers_

# Predict the cluster for each data point
labels = kmeans.predict(dataset)

# Plot the clustered points with different colors for each cluster
plt.scatter(dataset[:, 0], dataset[:, 1], c=labels, cmap='viridis', marker='o')

# Plot the centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=100, label='Centroids')

plt.legend()
plt.title('K-means Clustering with 3 Clusters')
plt.show()
```
