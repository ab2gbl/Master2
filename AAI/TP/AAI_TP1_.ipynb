```python
from PIL import Image
```

```python
def image_function(path):
    image=Image.open(path)
    original_image= image.copy ()
    grey_image=image.convert("L")
    pixels = list(grey_image.getdata())
    n_pixels=len(pixels)
    n_pblancs=pixels.count(255)
    n_pnoirs=pixels.count(0)
    img_res=image.resize((100, 100))
    img_rot=image.rotate(45)
    largeur, hauteur = image.size
    pix=image.load()
    for x in range(largeur):
         for y in range(hauteur):
            if x < 10 or x >= largeur - 10 or y < 10 or y >= hauteur - 10:
                pix[x, y] = 0
    return original_image,grey_image,pixels,n_pixels,n_pblancs,n_pnoirs,img_res,img_rot,largeur,hauteur,pix,image
```

```python
img_o,img_g,pixels,n_pixels,n_pblancs,n_pnoirs,img_res,img_rot,largeur,hauteur,pix,img=image_function('/pic.jpeg')
```
